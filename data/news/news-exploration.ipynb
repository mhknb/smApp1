```python
from utils.transformer import  DataTransformer
from utils.dataset import Dataset
from metrics import efficacy_test
from sklearn.metrics import accuracy_score
```

```python
train_dset = Dataset(
    dataset_name="news",
    dataset_dir="../",
    subset="train",
    random_state=1000,
)

print(train_dset.data.shape)
train_dset.data.head()
```

```python
transformer = DataTransformer(
        discrete_encode="onehot",
        numerical_preprocess="standard",
        target="shares_binary"
)

train_dset = Dataset(
    dataset_name="news",
    dataset_dir="../",
    subset="train",
    data_frac=None,
    random_state=1000,
    return_filtered_cols=True
)

test_dset = Dataset(
    dataset_name="news",
    dataset_dir="../",
    subset="test",
    data_frac=None,
    random_state=1000,
    return_filtered_cols=True
)


test_data = test_dset.data
train_data = train_dset.data
transformer.fit(train_data, train_dset.cat_cols)

print(train_data.shape, test_data.shape)
print(f"# categorical columns: {len(train_dset.cat_cols)}")
print(f"features/label dim: {test_dset.get_dim()}")

f1_score = efficacy_test(fakedata=train_data, realdata=test_data, 
                         target_name="shares_binary", transformer=transformer)
acc_score = efficacy_test(fakedata=train_data, realdata=test_data, 
                          target_name="shares_binary", 
                          transformer=transformer, scorer=accuracy_score)

print(f"F1 Score: {f1_score} \nAcc Score: {acc_score}")
```

```python
from utils.misc import geometric_sequence
seed = 1000
subsets = geometric_sequence(start_value=20, 
                             common_ratio=2, 
                             size=10)
```

```python
target_name="shares_binary"

transformer = DataTransformer(
        discrete_encode="onehot",
        numerical_preprocess="standard",
        target=target_name
)

test_dset = Dataset(
        dataset_name="news",
        dataset_dir="../",
        subset="test",
        data_frac=None,
        random_state=seed,
        return_filtered_cols=False
)

test_data = test_dset.data
transformer.fit(test_data, test_dset.cat_cols)

for subset in subsets:
    
    train_dset = Dataset(
        dataset_name="news",
        dataset_dir="../",
        subset="train",
        data_frac=subset,
        random_state=seed,
        return_filtered_cols=False
        )

    
    train_data = train_dset.data
    
    f1_train = efficacy_test(fakedata=train_data, realdata=train_data, target_name=target_name, transformer=transformer)
    acc_train = efficacy_test(fakedata=train_data, realdata=train_data, target_name=target_name, transformer=transformer, scorer=accuracy_score)
    
    f1_test = efficacy_test(fakedata=train_data, realdata=test_data, target_name=target_name, transformer=transformer)
    acc_test = efficacy_test(fakedata=train_data, realdata=test_data, target_name=target_name, transformer=transformer, scorer=accuracy_score)

    print(f"subset: {subset}: Train/Test-F1: {f1_train:.3f}/{f1_test:.3f} Train/Test-Acc: {acc_train:.3f}/{acc_test:.3f}")
    
```
